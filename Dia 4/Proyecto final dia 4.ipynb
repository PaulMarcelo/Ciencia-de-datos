PREPARAR EL CONJUNTO DE DATOS DE ENTRENAMIENTO Y PRUEBA

```python
X_cat = df_uci_credit_card[['SEX', 'EDUCATION', 'MARRIAGE']]
X_cat 
```

```python
from sklearn.preprocessing import OneHotEncoder
OneHotEncoder = OneHotEncoder()
X_cat = OneHotEncoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)
X_cat
```

```python
X_numerical = df_uci_credit_card[['LIMIT_BAL', 'AGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',
                'BILL_AMT1','BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
                'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']]
X_numerical
```

```python
X = pd.concat([X_cat, X_numerical], axis=1)
X
```

```python
from sklearn.preprocessing import MinMaxScaler


X.columns = X.columns.astype(str)

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
y = df_uci_credit_card['default.payment.next.month']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

ENTRENAR UN MODELO CLASIFICADOR XG-BOOST

```python
import xgboost as xgb
model_xgb = xgb.XGBClassifier(learning_rate = 0.1, max_depth = 20, use_label_encoder = False)
model_xgb.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_predict = model_xgb.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("Accuracy {} %".format( 100 * accuracy_score(y_predict, y_test)))
```

```python
cm = confusion_matrix(y_predict, y_test)
sns.heatmap(cm, annot = True, fmt = 'd');
```

```python
print(classification_report(y_test, y_predict))
```
