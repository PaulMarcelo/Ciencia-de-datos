TAREA DE PROGRAMACIÓN #3: IDENTIFICAR LA IMPORTANCIA DE LAS CARACTERÍSTICAS Y PREPARAR LOS DATOS ANTES DEL ENTRENAMIENTO DEL MODELO

```python
X = telecom_df.drop(["class", "area_code", "phone_number"], axis = "columns") 
y = telecom_df["class"]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 150)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train.values.ravel())
```

```python
feat_scores= pd.DataFrame({"Fracción de las variables afectadas" : rf.feature_importances_},index = X.columns)
feat_scores= feat_scores.sort_values(by = "Fracción de las variables afectadas")
feat_scores.plot(kind = "barh", figsize = (10, 5))
sns.despine()
```

TAREA DE PROGRAMACIÓN #4: ENTRENAR Y EVALUAR UN CLASIFICADOR DE REGRESIÓN LOGÍSTICA

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model_LR = LogisticRegression()

model_LR.fit(X_train, y_train)
```

```python
y_predict = model_LR.predict(X_test)
```

```python
print(classification_report(y_test, y_predict))
# La precisión es la relación TP/(TP+FP)
# El recall (recuperación) es la relación entre TP/(TP+FN)
# La puntuación F-beta puede interpretarse como una media armónica ponderada de la precisión y la recuperación.
# donde una puntuación F-beta alcanza su mejor valor en 1 y su peor puntuación en 0.
```

OPORTUNIDAD DE PRÁCTICA #3 [OPCIONAL]: Imprime la matriz de confusión y comenta los resultados

```python
print(confusion_matrix(y_test, y_predict))
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
plt.title('Matriz de confusión')
plt.xlabel('Predicción')
plt.ylabel('Realidad')
plt.show()
# La matriz de confusión muestra el número de verdaderos positivos, falsos positivos, verdaderos negativos y falsos negativos.
```

TAREA DE PROGRAMACIÓN #5: ENTRENAR Y EVALUAR UN CLASIFICADOR DE MÁQUINA DE VECTORES SOPORTE

```python
from sklearn.calibration import CalibratedClassifierCV # Para la salida de la puntuación de probabilidad
from sklearn.svm import LinearSVC

model_svc = LinearSVC(max_iter=100000)
model_svm = CalibratedClassifierCV(model_svc)
model_svm.fit(X_train, y_train)
```

```python
y_predict = model_svm.predict(X_test)
```

```python
print(classification_report(y_test, y_predict))
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot = True)
```

TAREA DE PROGRAMACIÓN #6: ENTRENAR Y EVALUAR UN CLASIFICADOR DE BOSQUE ALEATORIO

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
```

# TAREA DE PROGRAMACIÓN #9: COMPARACIÓN DE LOS MODELOS DE CLASIFICACIÓN ENTRENADOS Y OBSERVACIONES FINALES

```python
model_LR.predict_proba(X_test) #probabilidades de predicción
# El primer elemento es la probabilidad de que la salida sea 0
# El segundo elemento es la probabilidad de que la salida sea 1
# La suma de ambas probabilidades es 1
```

```python
fpr1, tpr1, thresholds1 = metrics.roc_curve(y_test, model_LR.predict_proba(X_test)[:, 1])
```

```python
from sklearn.metrics import roc_curve

fpr1, tpr1, thresholds1 = roc_curve(y_test, model_LR.predict_proba(X_test)[:, 1],pos_label = 1)
fpr2, tpr2, thresholds2 = roc_curve(y_test, model_svm.predict_proba(X_test)[:, 1],pos_label = 1)
fpr3, tpr3, thresholds3 = roc_curve(y_test, model_rf.predict_proba(X_test)[:, 1],pos_label = 1)
#fpr4, tpr4, thresholds4 = roc_curve(y_test, model_knn.predict_proba(X_test)[:, 1],pos_label = 1)
fpr5, tpr5, thresholds5 = roc_curve(y_test, model_gnb.predict_proba(X_test)[:, 1],pos_label = 1)
```

```python
from sklearn.metrics import roc_auc_score

auc_score1 = roc_auc_score(y_test, model_LR.predict_proba(X_test)[:, 1])
auc_score2 = roc_auc_score(y_test, model_svm.predict_proba(X_test)[:, 1])
auc_score3 = roc_auc_score(y_test, model_rf.predict_proba(X_test)[:, 1])
auc_score4 = roc_auc_score(y_test, model_gnb.predict_proba(X_test)[:, 1])
print("AUC score LR: ", auc_score1)
print("AUC score SVM: ", auc_score2)
print("AUC score RF: ", auc_score3)
print("AUC score GNB: ", auc_score4)
```

```python
plt.plot(fpr1, tpr1,linestyle="--", color = "orange", label = "LR (AUC = %0.2f)" % auc_score1)
plt.plot(fpr2, tpr2,linestyle="--", color = "red", label = "SVM (AUC = %0.2f)" % auc_score2)
plt.plot(fpr3, tpr3,linestyle="--", color = "green", label = "RF (AUC = %0.2f)" % auc_score3)
plt.plot(fpr5, tpr5,linestyle="--", color = "black", label = "GNB (AUC = %0.2f)" % auc_score4)

plt.title('Características del operador receptor (ROC)')
plt.xlabel('Tasa de falsos positivos')
plt.ylabel('Tasa de verdaderos positivos')

plt.legend(loc = 'best')
plt.savefig('ROC', dpi = 300)
plt.show()
```
